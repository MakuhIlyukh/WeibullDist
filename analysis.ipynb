```python
def cum_seconds(sub_run):
    return np.cumsum([x.iter_time for x in sub_run['res']], dtype=np.int64) / 10**9


def metrics(subrun):
    return np.array([x.metric for x in subrun['res']])

def lls(subrun):
    return np.array([x.loss for x in subrun['res']])

def plot_all_subruns(run_res):
    for sub_run in run_res:
        plt.plot(cum_seconds(sub_run), metrics(sub_run), alpha=0.1)


# plt.figure(figsize=(20, 20))
# for i, (key, value) in enumerate(meths.items()):
#     plt.subplot(3, 3, i + 1)
#     # adding space between subplots
#     # plt.subplots_adjust(hspace=0.5)
#     plot_all_subruns(value["results"][:1000])
#     plt.title(key)
#     plt.ylim(0, 1)
#     plt.xlim(0, 0.1)
#     plt.xlabel("Время (секунды)")
#     plt.ylabel("R2 Score")
```
